```python
#Import Pandas
import pandas as pd
import numpy as np
```

# Loading Olympic edition DataFrame

```python
#create file path
file_path = "datasets/medals/Summer Olympic medalists 1896 to 2008 - EDITIONS.tsv"
```

```python
#load Dataframe from file_path : editions
editions = pd.read_csv(file_path, sep='\t')
editions.head()
```

```python
#Extract the relevant columns
editions = editions[['Edition', 'Grand Total', 'City', 'Country']]
editions.head()
```

# Loading IOC Codes Dataframe


```python
ioc_path = "datasets/medals/Summer Olympic medalists 1896 to 2008 - IOC COUNTRY CODES.csv"
ioc_codes = pd.read_csv(ioc_path)
ioc_codes.head()
```

```python
#extract the relevant columns: 'Country' and 'NOC'
ioc_codes = ioc_codes[['Country', 'NOC']]
ioc_codes.head()
```

```python
ioc_codes.tail()
```

# BUILDING MEDALS DATAFRAME

```python
#create empty dictionary medals_dict
medals_dict = {}
```

```python
file_path = "datasets/medals/Summer Olympic medalists 1896 to 2008 - ALL MEDALISTS.tsv"
all_medalist = pd.read_csv(file_path, sep='\t', header=3)
all_medalist.columns = all_medalist.iloc[0]
#all_medalist = all_medalist.reindex(all_medalist.index.drop(0))
all_medalist.drop(0, inplace=True)
all_medalist.reset_index(inplace=True)
all_medalist.head()
```

```python
#create medalist by year - is stuff 
for group, name in all_medalist.groupby('Edition'):
    file_name = "datasets/medals/summer_%s_.csv" % group
    all_medalist[all_medalist['Edition'] == group].to_csv(file_name, index=False)
```

```python
# Extract the revelant columns
medals = all_medalist[['Athlete', 'NOC', 'Medal', 'Edition']]
medals.head()
```

# Counting medals by country/edition in a pivot table

```python
#construct the pivot_table: medal_counts
medal_counts = medals.pivot_table(index='Edition', columns='NOC', values='Athlete', aggfunc='count')
medal_counts.head()
```

# Computing fraction of medals per Olympic edition

```python
#Set Index Of Totals
totals = editions.set_index('Edition')
totals.head()
```

```python
# Reassing totals ['Grand Total'] : Total
totals = totals['Grand Total']
totals.head()
```

```python
#Divide Medal_Counts by totals : Count
fractions = medal_counts.divide(totals.values, axis='rows')
fractions.head()
```

# Computing percentage change in fraction of medals won

```python
mean_fractions = fractions.expanding().mean()
mean_fractions.head()
```

```python
fractions_change = mean_fractions.pct_change() * 100
fractions_change.head()
```

```python
fractions_change = fractions_change.reset_index()
fractions_change.head()
```

# Building Hosts DataFrame

```python
#LEFT EDITIONS AND IOC_CODES
hosts = pd.merge(editions, ioc_codes, how='left')
#pd.merge(ioc_codes, editions[['Edition', 'Country']], how='right', on='Country')
hosts.head()
```

```python
#extract relevant columns and set index
hosts = hosts[['Edition', 'NOC', 'City']].set_index('Edition')
hosts.head()
```

```python
hosts.loc[hosts.NOC.isnull()]
```

```python
## Fix missing 'NOC' values of hosts
hosts.loc[1972, 'NOC'] = 'FRG'
hosts.loc[1980, 'NOC'] = 'URS'
hosts.loc[1988, 'NOC'] = 'KOR'
```

```python
#RESET INDEX OF HOSTS
hosts.reset_index(inplace=True)
```

### Reshaping for analysis


```python
fractions_change.head()
```

```python
reshaped = pd.melt(fractions_change, id_vars='Edition', value_name='change')
reshaped
```

```python
#Extract row of reshaped where 'NOC' == "CHN"
chn = reshaped.loc[reshaped['NOC'] == 'FRA']
chn
```

### Merging to compute influence


```python
reshaped.head()
```

```python
hosts.head()
```

```python
# Merge reshaped and hosts: merged
reshaped['Edition'] = reshaped['Edition'].astype('int64')
merged = pd.merge(reshaped, hosts, how='inner')
merged
```

```python
# Set Index of merged and sort it: influence
influence = merged.set_index('Edition').sort_index()
influence
```

# Plotting influence of host country


```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Extract influence['Change']: change
change = influence['change']
```

```python
#make a bar plot of change
ax = change.plot(kind='bar')
#customize the plot to improve readability
ax.set_ylabel("Change of host Country Medal Count%")
ax.set_title("Is there a Host Country advantage?")
ax.set_xticklabels(influence['City'])
plt.show()
```
